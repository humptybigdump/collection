```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Funktionen u und utilde
def u(lam, y):
    return ((2000-y[0])**lam[0]) *((15-y[1])**lam[1])

def utilde(lam, y):
    return lam[0]*np.log(2000-y[0]) + lam[1]*np.log(15-y[1])
```

```python
# Datenpunkte (Fahrr√§der)

f1 = np.array([450, 900, 1300, 1250, 1600, 1750, 2000]) # preis
f2 = np.array([14, 13, 15, 10.5, 7, 12, 7]) # gewicht
# f1 = np.array([4,7,12,11,17,18,20]) # preis
# f2 = np.array([14,13,15,10,5,14,5]) # gewicht

# lambda
#lam = np.array([0.5,0.5])
lam= np.array([0.25,0.75])

omega = np.array([max(f1), max(f2)])
                 
```

a),b) - ($\lambda$ anpassen)

```python
# Plotdaten, als Gitter
y1 = np.arange(0,omega[0],0.1)
y2 = np.arange(0,omega[1],0.1)
y1, y2 = np.meshgrid(y1,y2)

# Funktionswerte
z = u(lam, [y1,y2])
z_trans = utilde(lam, [y1,y2])
```

```python
# Plot: 

# level f√ºr die H√∂henlinien
l = np.array([])
for i in range(7):
    l = np.append(l,u(lam, [f1[i],f2[i]]))
levels = np.unique(np.sort(l))

    
# plot
cs = plt.contour(y1, y2, z, levels)
plt.colorbar(cs)

plt.scatter(f1,f2, marker='x')
for i in range(7):
    plt.text(f1[i]+.2, f2[i]+.2, '$f(x^'+str(i+1)+')$', fontsize=9)

# Omega
plt.plot(np.linspace(0,omega[0],7),np.ones(7)*omega[1], color = 'black', linestyle='--')
plt.plot(np.ones(7)*omega[0], np.linspace(0,omega[1],7), color = 'black', linestyle='--')

plt.title('$u_{CD}$, $\\lambda =$'+str(list(lam))) #, $\lambda='+str(list(lam))+

plt.xlim([300,2100])
plt.ylim([5, 16])
plt.show()
```

```python
# Plot: 

# level f√ºr die H√∂henlinien
l = np.array([])
for i in range(7):
    l = np.append(l,utilde(lam, [f1[i],f2[i]]))
levels = np.unique(np.sort(l))
levels = np.delete(levels, 0) 

    
# plot
plt.contour(y1,y2,z_trans, levels)
plt.colorbar()
plt.scatter(f1,f2, marker='x')
for i in range(7):
    plt.text(f1[i]+.2, f2[i]+.2, '$f(x^'+str(i+1)+')$', fontsize=9)
plt.plot(np.linspace(0,omega[0],7),np.ones(7)*omega[1], color = 'black', linestyle='--')
plt.plot(np.ones(7)*omega[0], np.linspace(0,omega[1],7), color = 'black', linestyle='--')
plt.title('$\\tilde{u}_{CD}$, $\\lambda='+str(list(lam))+'$')

plt.xlim([300,2100])
plt.ylim([5, 16])
plt.show()
```
