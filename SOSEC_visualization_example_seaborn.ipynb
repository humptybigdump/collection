```python
import pandas as pd 

df = pd.read_csv("here, enter path to SOSEC dataset file")
```

Research question: How does the use of social networks relate to trust in the government in persons with different education levels and gender?


Hypothesis 1: More social media use is associated with lower trust.

Hypothesis 2: Higher education is associated with higher trust.

Hypothesis 3: Gender makes a difference.

```python
import seaborn as sns

df = df[df["F7a"].isin([1,2])]

#define subgroups for visualization
df.loc[df["F7g"]<=4, 'education'] = 'lower education'
df.loc[df["F7g"]>4, 'education'] = 'higher education'

print('persons with lower education:', len(df[df["F7g"]<=4]))
print('persons with higher education:', len(df[df["F7g"]>4]))

sns.relplot(
    data=df, kind="line",
    y="F5A5_1", x="F5aA4_1", col="F7a",
    hue="education", style="education",
)
```

H1: True

H2: True

H3: False
