- **Aufgabe 4**: Erstelle zwei Vektoren: Einer mit 20 zufälligen Integern zwischen 0 und 10, einer mit 10 zufälligen Werten zwischen 0 und 5. Entferne alle Werte vom ersten Vektor, welche auch im zweiten Vektor enthalten sind.

```python
# Aufgabe 3

# erstelle zufällige Werte
a = np.random.rand(20)
print(a)
# erstelle zufällige Indices
idx = np.random.randint(0, 20, 10)
# doppelte Werte vermeiden
idx = np.unique(idx)

# Rechenoperation durchführen
b = np.copy(a)
b[idx] += 1
print(b)

# checke die Berechnung
print(b - a)
```

```python
# Aufgabe 4

# erstelle Werte
a = np.random.randint(0, 10, 20)
print(a)
b = np.random.randint(0, 5, 10)
print(b)

idx = np.isin(a, b)
print(idx)

c = a[~idx]
print(c)
```

## Übung 3

Erweiterte Funktionalität:

- **Aufgabe 1**: Erstelle einen Vektor mit 1000 zufälligen ganzen Zahlen zwischen 0 und 100. Finde den häufigsten Wert in diesem Vektor.
- **Aufgabe 2**: Erstelle einen Vektor mit 10 Werten der Sinusfunktion im Intervall [0 und $\pi$]. Erweitere diesen Vektor um das Zehnfache durch lineare Interpolation.

- **Aufgabe 3**: Erstelle einen Vektor mit 1000 zufälligen Werten. Berechne den gleitenden Durchschnitt über 10 Werte.

```python
# Aufgabe 1

a = np.random.randint(0, 100, 1000)

# zähle jeden Wert
counts = np.bincount(a)
print(counts)
idx = np.argmax(counts)
print(counts[idx])
```

```python
# Aufgabe 2

x1 = np.linspace(0, np.pi, 10)
y1 = np.sin(x1)

# x-Vektor generieren
x2 = np.linspace(0, np.pi, 40)

# Werte interpolieren
y2 = np.interp(x2, x1, y1)
print(y2)
```

```python
# Aufgabe 3
a = np.random.rand(1000)

# eine Funktion für den gleitenden Durchschnitt
def runningMean(x, N):
    y = np.zeros((len(x),))
    for ctr in range(len(x)):
         y[ctr] = np.sum(x[ctr:(ctr+N)])
    return y/N

# Funktion anwenden
b = runningMean(a, 10)
print(b)
```
