# Diversity-multiplexing tradeoff

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.dpi'] = 300
Nr = 8
Nt = 8

r = np.arange(0, min(Nt, Nr)+1, 1) # +1 to account for arange excluding the endpoint
d_opt = (Nt - r) * (Nr - r)

plt.plot(r, d_opt)
plt.title("Diversity-multiplexing tradeoff")
plt.xlabel("r")
plt.ylabel("d_opt")
plt.show()
```
