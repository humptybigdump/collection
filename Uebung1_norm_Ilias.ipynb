## Vergleiche Sch√§tzung und Berechnung der Norm

Die Sch√§tzung basiert auf einer Folge von Realisationen $x_1, \dots, x_n$
\[
    (\frac{1}{n} \sum_{i=1}^n |x_i|^p)^{1/p}
\]

```python
from Uebung1_norm import histogram_of_norm_samples, compute_standard_normal_norm

norm_rv = stats.norm()
m = 10000

ps = [1, 2, 10, 30]

fig, axes = plt.subplots(1,len(ps))

for i, p in enumerate(ps):
    histogram_of_norm_samples(rv=norm_rv, p=p, m=m, ax=axes[i])
    axes[i].vlines(x=compute_standard_normal_norm(p=p), ymin=axes[i].get_ylim()[0], 
                   ymax=axes[i].get_ylim()[1], linewidth=2, color='g')
    axes[i].set_title(f'{p}-Norm')
    
fig.tight_layout()
```
