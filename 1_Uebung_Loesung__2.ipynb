## Übung 1

Ausbreitung einer Welle

$\varphi(x,t) = A exp(-\nu k^2 t) \cos(k(x-ct))$

a) Definition x Vektor mit N Einträgen im Intervall [0,1] 

```python
# Import der wichtigen Packages
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Definition Vektor x
N = 101
x = np.linspace(0,1,N)
```

b) Evaluierung der Wellenausbreitung für A=1, k=2$\pi$, $\nu$=1 und c=1

```python
# Definition der Parameter
A  = 1
k  = 2*np.pi
nu = 1
c  = 1

# Anzahl Zeitschritte
Nt = int(5/0.05+1)
# Zeitvektor
t  = np.linspace(0,5,Nt)
```

```python
#Enable interactive plot
%matplotlib notebook
%matplotlib notebook

from matplotlib import animation, rc
```

```python
fig,ax = plt.subplots()

line, = ax.plot([],[])

ax.set_xlim(0, 1)
ax.set_ylim(-1, 1)
```

```python
def animate(i):
    phi = A*np.exp(-nu*k**2*i*0.05)*np.cos(k*(x-c*i*0.05))
    line.set_data((x, phi))
    return (line,)
```

```python
anim = animation.FuncAnimation(fig, animate, frames=Nt, interval=200)
plt.show()
```
