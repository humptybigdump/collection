```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, norm, binom, poisson
from scipy.stats import gaussian_kde
```

## 1.

Plot of the power function:

a) Uniform distribution

```python
# Parameters
n = 100
alpha = 0.05
b_0 = 4
beta = np.arange(3, 7.01, 0.01)

# (1-alpha) quantile of the standard normal distribution
c = norm.ppf(1 - alpha)

# Compute the power function
G = np.zeros(len(beta))
for i in range(len(beta)):
    G[i] = 1 - norm.cdf(c*b_0/beta[i] + np.sqrt(3*n)*(b_0/beta[i]-1))

# Plot the power function
plt.figure()
plt.plot(beta, G)
plt.xlabel(r'$\beta$', fontsize=12)
plt.ylabel(r'$G(\beta)$', fontsize=12)
plt.title('Power function (one-sided test)')
plt.grid()
plt.show()
```

b) Poisson distribution:

```python
# Parameters
n = 100
alpha = 0.05
mu_0 = 4
mu = np.arange(1, 7.01, 0.01)

# (1-alpha) quantile of the standard normal distribution
c = norm.ppf(1 - alpha/2)

# Compute the power function
F = np.zeros(len(mu))
for i in range(len(mu)):
    term1 = (mu_0 - mu[i] + c * np.sqrt(mu_0 / n)) / np.sqrt(mu[i] / n)
    term2 = (mu_0 - mu[i] - c * np.sqrt(mu_0 / n)) / np.sqrt(mu[i] / n)
    F[i] = 1 - norm.cdf(term1) + norm.cdf(term2)

# Plot the power function
plt.figure()
plt.plot(mu, F)
plt.xlabel(r'$\mu$', fontsize=12)
plt.ylabel(r'$G(\mu)$', fontsize=12)
plt.title('Power function (two-sided test)')
plt.grid()
plt.show()
```
