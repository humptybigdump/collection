# BDA Challenge: Data Analysis & Visualisation

## Load Necessary Libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
import json
from pandas.io.json import json_normalize
```

## Load Data

**DataFrame:** 2-dimensional labeled data structure with columns of potentially different types (e.g., str, int)

```python
df = pd.read_json('response_BDA.txt', lines=True)
df
```

```python
df_garbage = pd.DataFrame(df[1][0])
df_garbage
```

## Normalize Nested Dictionary

**Dictionary:** used to store data values in key:value pairs.

**Nested Dictionary:** a dictionary inside a dictionary

```python
dict_garbage = df_garbage.decoded_data[0]
keys_garbage = dict_garbage.keys()
values_garbage = dict_garbage.values()
values_garbage

df_decoded_data = json_normalize(df_garbage.decoded_data)
df_decoded_data
```

## Change str to int

```python
df_decoded_data["sensor_data.Height 1 (cm)"] =  df_decoded_data["sensor_data.Height 1"].str.replace("cm","").astype(int)
df_decoded_data
```

## Test Plotting

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]

plt.plot(x, y)
plt.plot(x, z)

plt.title("test plot")

plt.xlabel("x")
plt.ylabel("y and z")

plt.legend(["this is y", "this is z"])

plt.show()
```

## Plotting "Sensor Height" over "Created at"

```python
#Putting columns of data frame into lists
created_at = df_garbage.created_at.tolist()
sensor_height_1 = df_decoded_data["sensor_data.Height 1 (cm)"].tolist()

#Choosing Values for Plotting
n = 50
x = created_at[0:n]
y = sensor_height_1[0:n]

#Plotting
plt.figure(figsize=(10,6))
plt.plot(x, y)

plt.xticks(fontsize=8, rotation=90)
plt.yticks(fontsize=10, fontweight='bold')

plt.title("Time & Level")
plt.xlabel("Created at")
plt.ylabel("Sensor Height")

plt.show()
```
