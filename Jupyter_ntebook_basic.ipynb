### Graphical representation of data

```python
# Generate random integers and uniform random numbers
z = np.random.randint(1, 51, (20, 1))  # generates 20 random integers between 1 and 50
y = np.random.rand(20, 1)  # generates 20 random uniform numbers

# Create the first plot
plt.figure(1)  # create the first figure window
plt.plot(z, y, '.')  # plot data points
```

```python
# Generate a cubic function
x = np.arange(1, 21).reshape(-1, 1)  # creates a column vector from 1 to 20
t = x**3 - 5  # cubic function

# Create the second plot
plt.figure(2)  # create the second figure window
plt.plot(x, t)
```

```python
# Generate sine functions
x = np.arange(0, 2 * np.pi, np.pi / 100)  # range from 0 to 2π with a step size
y1 = np.sin(x)
y2 = np.sin(x - 0.25)
y3 = np.sin(x - 0.5)

# Create the third plot
plt.figure(3)
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, '--', label='sin(x - 0.25)')
plt.plot(x, y3, ':', label='sin(x - 0.5)')
plt.xlabel('x-axis')  # label for x-axis
plt.ylabel('y-axis')  # label for y-axis
plt.title('Sine')  # title of the plot
plt.legend()  # show legend
# plt.grid(True)  # uncomment to enable grid
```

```python
# Linear regression: model y = b_0 + b_1*z + e
X = np.hstack((np.ones((20, 1)), z))  # create design matrix with ones
b_hat = np.linalg.inv(X.T @ X) @ X.T @ y  # calculate coefficients
y_hat = X @ b_hat  # predicted values

# Create the fourth plot
plt.figure(4)
plt.plot(z, y, 'o', label='Data Points')  # plot original data
plt.plot(z, y_hat, 'r', label='Regression Line')  # plot regression line
plt.xlabel('z-axis')
plt.ylabel('y-axis')
plt.title('Linear Regression')
plt.legend()
plt.show()  # show plots
```

#### Example: Quadratic function

```python
x = np.arange(-10, 10.5, 0.5)  # range from -10 to 10 with step 0.5
f = 0.7 * x**2 + x + 12  # quadratic function

# Create the fifth plot
plt.figure(5)
plt.plot(x, f)  # plot the quadratic function

# Numerical optimization
result = minimize_scalar(lambda x: 0.7 * x**2 + x + 12, bounds=(-10, 10), method=None)
min_x = result.x  # found minimum
f_min = result.fun  # minimum value of the function
print(f"Minimum value of the function is {f_min} at x = {min_x}")  # print result

plt.show()  # show plots
```
