# Aufgabe 3
Gegeben sei das Signal $x(t) = \cos\left(2\pi f_\mathrm{c} t\right)$ mit $f_\mathrm{c}=9\,\mathrm{Hz}$. In der Zeit von $t_0=0\,\mathrm{s}$ bis $t_1=0.5\,\mathrm{s}$ wird es mit der Abtastfrequenz $f_\mathrm{a}=27\,\mathrm{Hz}$ zu $x[n]$ abgetastet. Schreiben Sie eine Simulation mit folgenden Schritten:
- Berechnen Sie $x[n]$ und vergleichen Sie dieses grafisch mit $x(t)$ (näherungsweise).
- Berechnen Sie die DFT $X[k]$ von $x[n]$. Stellen Sie diese grafisch über der Frequenz in Hz dar. 

```python
# importing
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# showing figures inline
%matplotlib inline
matplotlib.rc('figure', figsize=(30, 8) )
```

# Sampling

```python
fa = 27 # Hz, Samplingfrequenz
ta = 1/fa # Abtastzeit
t0 = 0
t1 = 0.5 # s
tn = np.arange(t0,t1,ta) # Zeitvektor
n = np.arange(tn.size) 

fc = 9 
x = np.cos(2*np.pi*fc*tn)

# "Zeiktontinuierlich":
t = np.linspace(t0,t1,1000)
x_kont = np.cos(2*np.pi*fc*t)
```

# Plot x(t) vs x[n]

```python
plt.plot(t,x_kont, 'k-', label="x(t)")
plt.stem(tn,x, label="x[n]")
plt.legend()
```

# DFT & Plot

```python
X = np.fft.fft(x)
f = np.fft.fftfreq(tn.size, d=ta)

plt.stem(f,np.abs(X))
print(f"N = {tn.size}")
```

# Aussagekräftigere DFT

```python
X = np.fft.fft(x, n=12)
f = np.fft.fftfreq(12, d=ta)

plt.stem(f,np.abs(X))
```
