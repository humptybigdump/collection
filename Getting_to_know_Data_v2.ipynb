## Stationarity Test for RETURNS: Augmented Dickey Fuller Test

$$
\\
$$

**Custom-made ADF Function for Pandas DataFrame $r_d$**

```python
#ADF Test
#
def ADF(r):
    #input: pd.DataFrame
    #output: p-values and print of a string
    
    output = "Data is stationary" #initialize
    
    pvle = pd.DataFrame()
    
    for i in range(0,r.shape[1]):
        adf_test = adfuller(r.iloc[:,i])
        pvle.loc[i,0] = adf_test[1]
        
        if adf_test[1] > 0.05:
            output = "At least one column of input data is NON-STATIONARY"
        
        
    print(output)
    
    return pvle
    
```

$$
\\
$$

**Is Return Panel $r_d$ Stationary?**

```python
#Stationarity Test on r_d
pvl_r = ADF(r_d) 
```

```python
#max p-vle
pvl_r.max()   #clearly 0 evidence for non-stationarity in return data
```

### Custom-Made Function to Visualize Outiers in Return Panel $r_d$ 

```python
def print_z(r,d,p):
    #input: d threshold
    #input: r return panel
    #input: p sec for pause
    
    z        = stats.zscore(r)
    tickers  =  r.agg(['mean']).T.index
    
    counter = 0
    for i in tickers:
        print(i)
        x = r.index
        y = z[:,counter]
        
        #record movements that are larger than threshold d
        filter = np.abs(y)>d
        
        #plots outliers as a function of time
        plt.scatter(x[filter], y[filter] )
        plt.pause(p)
        counter +=1
     
  
    
```

**Visualize 10-std Returns for $r_d$**

```python
print_z(r_d,10,1)
```

## Risk Decomposition, i.e. Total Risk = Systematic Risk + Idiosyncratic Risk

**Note: Annualized and in decimals**

**correlation of assets with market**

```python
InvOpp["corr_1/N"]   = r_d.corr().iloc[:,r_d.shape[1]-1] # Get correl(r_i,r_1/N)
```

**total risk**

```python
InvOpp["Total.Risk"] = InvOpp["std"]**2
InvOpp["Total.Risk"] = InvOpp["Total.Risk"] * 252
```

**systematic risk, i.e. cov(r_i, r_MKT) in CAPM theory**

```python
InvOpp["Systm.Risk"] = InvOpp["std"].iloc[InvOpp.shape[0]-1] * InvOpp["std"] * InvOpp["corr_1/N"] 
InvOpp["Systm.Risk"] = InvOpp["Systm.Risk"] * 252
```

**systematic risk relative to total risk**

```python
InvOpp["Systm.Risk.Ratio"] =  InvOpp["Systm.Risk"] /InvOpp["Total.Risk"]  
```

**CAPM Beta, i.e. normalized systematic risk**

```python
InvOpp["Beta"]       = InvOpp["Systm.Risk"] / InvOpp["Systm.Risk"].loc["1/N"] #CAPM beta_i := systemRisk_i / systemRisk_MKT
```

**Idiosyncratic Risk**

```python
InvOpp["Idiosync.Risk"] = InvOpp["Total.Risk"] - InvOpp["Systm.Risk"]
InvOpp["Idiosync.Risk"] = InvOpp["Idiosync.Risk"] * 252
```

**How much of a firms' risk is diversifiable?**

```python
InvOpp["Idiosync.Risk.Ratio"] = 1-InvOpp["Systm.Risk.Ratio"]
```

### Observations and Implications for PCA

Ma, Zabarus (2011, Journal of Computational Physics 230) write on page 7312 "The Kâ€“L expansion based stochastic model reduction scheme constructs the closest linear subspace of the high-dimensional input space. In other words, it only **preserves the mean and covariance** of the random field, and therefore is suitable for multi- Gaussian random fields. But most of the random samples contain essential non-linear structures, e.g. higher-order statistics. This directly translates into the fact that the standard linear Kâ€“L expansion tends to over-estimate the actual intrinsic dimen- sionality of the underlying random field. Hence, one needs to go beyond a linear representation of the high-dimensional in- put space to accurately access the effect of its variability on the output variables."

The authors continue on that page "Both problems of the Kâ€“L expansion and the non-linear model reduction algorithm in [29] can be solved with kernel prin- cipal component analysis (KPCA), which is a nonlinear form of PCA [31,32]. KPCA has proven to be a powerful tool as a non- linear feature extractor of classification algorithm [31], pattern recognization [33], image-denosing [34] and statistical shape analysis [35]. The basic idea is to map the data in the input space to a feature space F through some nonlinear map U, and then apply the linear PCA there. Through the use of a suitably chosen kernel function, the data becomes more linearly related in the feature space F. In the context of stochastic modeling, there are two pioneering works [36,37]. In [36], KPCA was used to construct the prior model of the unknown permeability field and then gradient-based algorithm is used to solve the his- tory-matching problem. However, the random expansion coefficients of the linear PCA in the feature space are assumed i.i.d. standard normal random variables. This choice clearly does not capture the statistical information from the data."

Their conclusion contains the following: ""**Linear PCA has been proved to be effective in various scientific areas. In general, linear PCA is still the first choice for model reduction. KPCA may be less effective than linear PCA when data is linearly-related.**

## Plotting Realized PDF of Daily Returns 
$$
\\
$$

**Custom-Made Function for Histogram of Normalized (z) Realized Returns, incl. JB-Test**

```python
def print_pdf(r,b,p):
    #input: b #of bins in histogram
    #input: r return panel
    #input: p sec for pause
    
    
    InvOpp  =  r.agg(['mean', 'std']).T 
    tickers = InvOpp.index
   
    
    counter = 0
    for i in tickers:
        print(i)
        #JB test
        JB_test = stats.jarque_bera(r.iloc[:,counter]) #JB test
        if JB_test[1] >0.1: #p-value of 10%, Generous for conclusing r can be modelled via N()
            print('Gaussian Density, N()')
        else:
            print('Non-Gaussian Density')
        #plot    
        x = r.iloc[:,counter]
        plt.hist(stats.zscore(x), bins=b, density=True )
        plt.xlabel("z(r_d)")
        plt.ylabel("hist(r_d)")
        plt.axis('tight')
        plt.show()
        plt.pause(p)
        counter +=1
```

**Visualize PDF of Normalized (z) Realized Returns**

```python
print_pdf(r_d, 20, 1)
```

### Option-implied Beta

https://papers.ssrn.com/sol3/papers.cfm?abstract_id=1301437

**Abstract:**

We use forward-looking information from option prices to estimate option-implied correlations and to construct an option-implied predictor of factor betas. With our implied market betas, we find a monotonically increasing risk-return relation, not detectable with standard rolling-window betas, with the slope close to the market excess return. Our implied betas confirm a risk-return relation consistent with linear factor models, because, when compared to other beta approaches: (i) they are better predictors of realized betas, and (ii) they exhibit smaller and less systematic prediction errors.
