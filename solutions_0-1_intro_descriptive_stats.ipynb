**1.2** What is the difference between mean and median and when would you apply each?

__Solution__: The mean is the arithmetic mean of a sample and is sensitive to extreme values. The median is the central value of samples sorted increasingly. It is robust in cases where extreme values occur and can be interpreted as the typical value of a distribution.

**1.4** Plot the average number of cyclists over the course of a week (monday, tuesday, ...). Tipp: Use groupby

```python
df = pd.read_csv('Fremont.csv', index_col='Date', parse_dates=True)
df = df.rename(columns={"Fremont Bridge Sidewalks, south of N 34th St": "total", 
                        "Fremont Bridge Sidewalks, south of N 34th St Cyclist East Sidewalk": "east",
                        "Fremont Bridge Sidewalks, south of N 34th St Cyclist West Sidewalk": "west"}) 
```

```python
df.groupby(df.index.dayofweek).mean().plot()
_ = plt.xticks(ticks=[0,1,2,3,4,5,6],labels=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"])
```

**1.7** Compute the correlations at different temporal resolutions (daily, weekly, monthly). Tip: Use resample() or groupby()\

```python
X_daily = df['east'].resample('d').mean()
y_daily = df['west'].resample('d').mean() 
X_weekly = df['east'].resample('w').mean()
y_weekly = df['west'].resample('w').mean() 
X_monthly = df['east'].resample('m').mean()
y_monthly = df['west'].resample('m').mean() 

r_daily, p_daily = pearsonr(X_daily,y_daily)
r_weekly, p_weekly = pearsonr(X_weekly,y_weekly)
r_monthly, p_monthly = pearsonr(X_monthly,y_monthly)

print(r_daily, r_weekly, r_monthly)
```

**1.8** Compute a linear regression between east and west for the monthly means.
Compute the correlations at different temporal resolutions (daily, weekly, monthly).

```python
X = df['east'].resample('m').mean()
y = df['west'].resample('m').mean()
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(X,y)

plt.plot(X,y, 'o', label='original data')
plt.plot(X, intercept + slope*X, 'r', label='fitted line')
plt.xlabel('east')
plt.ylabel('west')
plt.legend()
plt.show()
```

```python
# yes, the two samples differ significantly (p<0.05) 
t,p=scipy.stats.ttest_ind(X,y)
p<0.05
```
