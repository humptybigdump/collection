```python
class LKW:
    def __init__(self, name, max_ladung, ladung, reichweite, ort_A):
        self.name = name
        self.max_ladung = max_ladung
        self.ladung = 0
        self.reichweite = reichweite
        self.ort_A = ort_A
        
    def fahren(self):
        if self.reichweite <= 100:
            print(f"Reichweite von {self.name} ist nicht ausreichend.")
            return False
        else:
            if self.ort_A: 
                self.ort_A = False
            else:
                self.ort_A = True
            self.reichweite -= 100
            return True
            
    def beladen(self, menge):
        if self.ladung > 0:
            print(f"{self.name} ist schon beladen.")
        else:
            if menge > self.max_ladung or menge < 0:
                print(f"{self.name} kann nicht mit dieser Menge beladen werden.")
            else:
                self.ladung += menge
                print(f"{self.name} wurde mit {menge} beladen.")
                return menge
    
    def abladen(self):
        if self.ladung <= 0:
            print(f"{self.name} ist leer")
            return 0
        else:
            print(f"{self.name} hat {self.ladung} abgeladen")
            ladung = self.ladung
            self.ladung = 0
            return ladung
    
    def tanken(self, sprit):
        self.reichweite += sprit
    
    def daten(self):
        print(f"Name: {self.name}, Ladung: {self.ladung}, Reichweite: {self.reichweite}, \
              Ort A: {self.ort_A}")
```

```python
holz_ort_A = 4000
holz_ort_B = 0

lkw_klein = LKW("lkw_small", 500, 0, 800, True)
lkw_gross = LKW("lkw_big", 2000, 0, 200, True)

while holz_ort_A > 0:
    # Dadurch wird vermieden, dass mehr Holz aufgeladen wird, als am Ort A vorhanden ist.
    ladung = min(500, holz_ort_A) 
    ladung_lkw_klein = lkw_klein.beladen(ladung)
    holz_ort_A -= ladung_lkw_klein
    
    ladung = min(2000, holz_ort_A)
    ladung_lkw_gross = lkw_gross.beladen(ladung)
    holz_ort_A -= ladung_lkw_gross

    ausreichend = lkw_klein.fahren()
    if not ausreichend:
        lkw_klein.tanken(300)
    
    ausreichend = lkw_gross.fahren()
    if not ausreichend:
        lkw_gross.tanken(300)

    holz_ort_B += lkw_klein.abladen()
    holz_ort_B += lkw_gross.abladen()

    print("HOLZ (Ort A)): ", holz_ort_A)
    print("HOLZ (Ort B)): ", holz_ort_B)

    lkw_klein.fahren()
    lkw_gross.fahren()
```
