```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(0,5,0.01)
x_eff = np.arange(0,2,0.01)

def f1(x):
    return np.sqrt(1+x)
def f2(x):
    return (x-2)**2+1

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(f1(x), f2(x), 'black')
ax1.plot(f1(x_eff), f2(x_eff),'red', label='$Y_{nd}$')
ax1.set_xlabel('$f_1$')
ax1.set_ylabel('$f_2$')
plt.legend()
plt.savefig('a3_ynd.png')
plt.show()
```
